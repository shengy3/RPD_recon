```python
def fill_array(arr):
    output = []
    
    
    for branch in arr.keys():
        #count = 0
        print(branch)
        tmp = array( 'd' )
        for i in arr[branch]:
            #if count > 10:
            #    break
            #print(i)
            tmp.append(i)
            
        output.append(tmp)
       
    return output
```

```python
def get_df(path, tree = 'tree', branch = []):
    file = uproot.open(path)
    if not branch:
        branch = file[tree].keys()
    df = file[tree].pandas.df(branch , flatten=False) 
    return df
```
